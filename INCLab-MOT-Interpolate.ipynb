# interpolate the raw data
- dataset_name = "raw csv file"
- dataset_name_without_comma = "csv file which deleted comma in certain value that has comma"
- result_name = "final csv file"



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

path = os.getcwd()
file_name = "\\INCLab_RawData\\"
path = path + file_name
```

```python
dataset_name =path+"0813_RawResult.csv"
dataset_name_without_comma ="0813_before_postProcessing_without_comma.csv"
result_name ="Result-0813-Fin.csv"
```

```python
df = pd.read_csv(dataset_name)
# pd.options.display.max_rows = 60
# pd.options.display.max_columns = 20
df.describe()
```

# chage the type of column
Values connected by commas (,) go into object types
Remove value after comma (,) and convert type to numeric form

```python
df_without_time = df.drop(['frameNum'], axis = 1)
df_without_time
```

```python
def remove_comma(x):

    if str(x).find(",") != -1:
        return int(str(x).split(",")[0])
    return x
```

```python
for i in df_without_time.columns:
    df_without_time[i] = df_without_time[i].apply(remove_comma)
df_without_time
```

```python
df_without_time.to_csv(dataset_name_without_comma) 
```

```python
df_without_time["frameNum"]=df["frameNum"]
df_without_time = df_without_time.reindex(columns=["frameNum","1_X", "1_Y","2_X", "2_Y" 
                                     ,"3_X", "3_Y" ,"4_X", "4_Y" 
                                     ,"5_X", "5_Y" ,"6_X", "6_Y" 
                                     ,"7_X", "7_Y" ,"8_X", "8_Y" 
                                     ,"9_X", "9_Y" ,"10_X", "10_Y" 
                                     ,"11_X", "11_Y" ,"12_X", "12_Y" 
                                     ,"13_X", "13_Y" ,"14_X", "14_Y" 
                                     ,"15_X", "15_Y" ,"16_X", "16_Y" ])
df_without_time
```

# Before interpolate

```python
print("**********")  
print(df_without_time.isna().sum())  
print("\n")  
```

```python
#change the type of column aftere delete comma

for column in df_without_time:
    df_without_time[column]=pd.to_numeric(df_without_time[column])
    #plt.plot(df_without_time["frameNum"],df_without_time[column])
    #plt.show()


df_without_time.describe()
```

```python
# add additional column named "nc1~nc16", which types are boolean that represent null or not
#nc = null cnt

df_nullornot = df_without_time.isnull()
df_nullornot=df_nullornot.drop(['frameNum'], axis = 1)
df_nullornot=df_nullornot.drop(['1_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['2_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['3_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['4_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['5_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['6_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['7_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['8_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['9_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['10_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['11_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['12_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['13_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['14_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['15_Y'], axis = 1)
df_nullornot=df_nullornot.drop(['16_Y'], axis = 1)

df_nullornot.rename(columns = {'1_X' : 'nc1', 
                              '2_X' : 'nc2', 
                              '3_X' : 'nc3', 
                              '4_X' : 'nc4',
                              '5_X' : 'nc5', 
                              '6_X' : 'nc6',
                              '7_X' : 'nc7', 
                              '8_X' : 'nc8',
                              '9_X' : 'nc9', 
                              '10_X' : 'nc10',
                              '11_X' : 'nc11', 
                              '12_X' : 'nc12',
                              '13_X' : 'nc13', 
                              '14_X' : 'nc14',
                              '15_X' : 'nc15', 
                              '16_X' : 'nc16',}, inplace = True)
```

# After interpolate

```python
df_without_time=df_without_time.interpolate(method ="spline",
              order =1,
            axis=0,
            limit_area="inside",
            limit_direction=None)
df_without_time=df_without_time.bfill(axis=0)
df_without_time = df_without_time.ffill(axis=0)

```

```python
df_without_time=pd.concat([df_without_time, df_nullornot],axis=1)
df_without_time
```

```python
df_without_time.to_csv(result_name)
```
