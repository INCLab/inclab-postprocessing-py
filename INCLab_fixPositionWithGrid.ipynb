# fixing position by Grid Table
- result_name = "csv file that only interpolated"
- df_result_raw = "csv file that done by Grid Table"
- steps = how big cells are


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

date ="0507" #0611 0514 0507
#result_name = "Result-0709-#2_enter_10min_late-Fin.csv"
#result_name = "Result-0604-#10-21120_25179-Fin.csv"
result_name ="Result-"+date+"-Fin.csv"
df_result_raw = "withGrid_"+date+".csv"
```

```python
df = pd.read_csv(result_name)

# pd.options.display.max_rows = 60
# pd.options.display.max_columns = 20
df_x=df[['1_X', '2_X',  '3_X', '4_X', '5_X',  '6_X', '7_X', '8_X', 
       '9_X', '10_X', '11_X',  '12_X', '13_X', '14_X', '15_X', '16_X']]
df_y=df[[ '1_Y',  '2_Y',  '3_Y', '4_Y', '5_Y',  '6_Y',  '7_Y', '8_Y',
       '9_Y','10_Y', '11_Y',  '12_Y', '13_Y', '14_Y', '15_Y',  '16_Y']]
```

```python
#Griding X position


#1920×1080
steps=2
x_grid =np.arange(0, 1920, step=steps)
y_grid =np.arange(0, 1080, step=steps)
# 이전 프레임의 값을 고정해두고, 그 값이 속해있던 그리드를 벗어나면 그 값으로 다시 초기화 하는 방식.
for i in df_x.columns:
    c=0
    g=0
    print(i)
    for index in range(df["1_X"].size): 
        if(index==0):
            for grid in x_grid:
                if (df_x[i][index]>=grid and df_x[i][index]<(grid+steps)):
                    c=df_x[i][index]
                    g=grid
                    break
        else:
            for grid in x_grid:
                if (df_x[i][index]>=grid and df_x[i][index]<(grid+steps)):
                    if (g==grid) :
                        df_x[i][index]=c
                        break
                    else :
                        c=df_x[i][index]
                        g=grid
                        break
```

```python

# Griding Y position

for i in df_y.columns:
    c=0
    g=0
    print(i)
    for index in range(df["1_Y"].size):
        if(index==0):
            for grid in y_grid:
                if (df_y[i][index]>=grid and df_y[i][index]<(grid+steps)):
                    c=df_y[i][index]
                    g=grid
                    break
        else:
            for grid in y_grid:
                if (df_y[i][index]>=grid and df_y[i][index]<(grid+steps)):
                    if (g==grid) :
                        df_y[i][index]=c
                        break
                    else :
                        c=df_y[i][index]
                        g=grid
                        break
```

```python
#concat X and Y position, also adding frameNum
result = pd.concat([df_x,df_y],axis=1)
result["frameNum"] = df["frameNum"]
result = result.reindex(columns=["frameNum","1_X", "1_Y","2_X", "2_Y" 
                                     ,"3_X", "3_Y" ,"4_X", "4_Y" 
                                     ,"5_X", "5_Y" ,"6_X", "6_Y" 
                                     ,"7_X", "7_Y" ,"8_X", "8_Y" 
                                     ,"9_X", "9_Y" ,"10_X", "10_Y" 
                                     ,"11_X", "11_Y" ,"12_X", "12_Y" 
                                     ,"13_X", "13_Y" ,"14_X", "14_Y" 
                                     ,"15_X", "15_Y" ,"16_X", "16_Y" ])
result[["nc1", "nc2", "nc3", "nc4", "nc5", "nc6", "nc7", "nc8", 
        "nc9", "nc10", "nc11", "nc12", "nc13", "nc14", "nc15", "nc16"]]= df[["nc1", "nc2", "nc3", "nc4", "nc5", "nc6", "nc7", "nc8", 
                                                                             "nc9", "nc10", "nc11", "nc12", "nc13", "nc14", "nc15", "nc16"]]
print(result)
result.to_csv(df_result_raw)
```
