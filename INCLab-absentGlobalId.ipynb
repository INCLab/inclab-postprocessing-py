# After interpolate then delete absent number in certain frameNum

0709 : **11415** 번 frame 부터 2 번 입장

0604 : **21120~25179** frame동안 10번 없었음


```python
import numpy as np
import pandas as pd

df= pd.read_csv("Result-0604-Fin.csv")
df
```

```python
start_frame=21120
end_frame= 25179

for index in df.frameNum:
    if index>=start_frame and index <= end_frame:
        df["10_X"][index] = np.nan
        df["10_Y"][index] = np.nan

df.to_csv("Result_0604_#10_21120_25179.csv")
```
