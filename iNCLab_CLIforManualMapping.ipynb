# CLI for manual mapping

* 여기서 저장이라 함은 csv 파일에 입력하는 것을 뜻한다.

1. 해당 csv 파일을 읽어 df에 올려둔다 
2. 번호, 시작 프레임, 종료 프레임을 입력하여 수정값을 반영한다
3. 업데이트 된 df를 저장한다. (저장버튼, 저장키 = "q")

df를 csv로 저장하면 그 csv를 읽어와야함 -> 오버헤드



```python
import numpy as np
import pandas as pd

date ="0507" #0611 0514 0507
#result_name = "Result-0709-#2_enter_10min_late-Fin.csv"
#result_name = "Result-0604-#10-21120_25179-Fin.csv"
result_name ="Result-"+date+"-Maually_Fin.csv"
df_result_raw = "withGrid_"+date+".csv"

df= pd.read_csv(df_result_raw)
df
```

```python

while (1):
    quitOrNum = input("if you want to quit please enter 'q' else 1~16 : ")
    if (quitOrNum == "q"):
        df.to_csv(result_name)
        print(df)
        print("saved")
        break;

    global_id=quitOrNum
    XorY =input("enter 1 for X, 2 for Y :")
    start_frame =int(input("enter 'start frame' to fix : ")) 
    end_frame=int(input("enter 'end frame' to fix : "))
    value=input("enter value to update : ")
    # df
    if (XorY == "1") :
        column =str(global_id)+"_X"
        print(df)
        for index in range(start_frame, end_frame+1):
            df.loc[df['frameNum'] == index,column]=value
    elif(XorY == "2") : 
        column =str(global_id)+"_Y"
        for index in range(start_frame, end_frame+1):
            df.loc[df['frameNum'] == index,column]=value
    print("--------!!!UPDATED!!!-----------")
```
